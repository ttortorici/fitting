```python
import fitting.capacitor as cap
import matplotlib.pylab as plt
import numpy as np
plt.rcParams['text.usetex'] = True


gap = np.linspace(1, 19, 100)
bare = cap.bare(gap, silica_constant=3.56)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6.5, 4.5))

ax.plot(gap, bare)
ax.scatter(5, 1.65)
ax.scatter(6, 1.53)
ax.grid()
ax.set_xlabel("gap width ($\\mathrm{{\\mu m}}$)")
ax.set_ylabel("capacitance (pF)")
fig.tight_layout()
```

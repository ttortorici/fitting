```python
import fitting.capacitor as cap
import matplotlib.pylab as plt
import numpy as np
plt.rcParams.update({
    "text.usetex": True,
    "text.latex.preamble": r"\usepackage{amsmath, textcomp}"
})

gap = np.linspace(1, 19, 100)
silica = np.array([[3.5], [3.7], [3.9]])
bare = cap.bare(gap, silica_constant=silica)
print(gap.shape)
print(silica.shape)
print(bare.shape)
print(len(silica))
print(bare[0].shape)
print(silica[2, 0])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6.5, 4.5))

for ii in range(len(silica)):
    ax.plot(gap, bare[ii], label=f"$\\varepsilon={silica[ii, 0]}$")
ax.scatter(5, 1.65, label="TT2 for old BTP film") # old BTP
ax.scatter(6.1, 1.469, label="TT5 for BTB 4") # BTB 4
ax.scatter(5.5, 1.53, label="TT5 for BTB 5") # BTB 5
ax.scatter(4, 1.71, label="TTX1 for GBA230")

ax.grid()
ax.legend()
ax.set_title(f"Bare Capacitance using Wu Model")
ax.set_xlabel("gap width ($\\text{{\\textmu}}\\text{{m}}$)")
ax.set_ylabel("capacitance (pF)")
fig.tight_layout()
fig.savefig("bare.png", dpi=600, bbox_inches="tight")
```
